```python
# So, how to save an image to an local folder?
# matplot library can be used to save an image to a local folder
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# A path, where the image will be save to, should be created. 
# The image to be saved is named as 'OneTower.jpg' 
img_path = r'C:\Temp\OneTower.jpg'

# Here is one method to save the image:
mpimg.imsave(img_path, BerkeleyTower)

# Here is another method to save the image. 
# You can comment out the above method and uncomment this one to test：
#plt.imsave(img_path, BerkeleyTower)
```

```python
# Now, how to convert a Colour image to Gray-scale image?
import matplotlib.pyplot as plt

# separate the R, G, and B channels
red_channel = BerkeleyTower[:,:,0]
green_channel = BerkeleyTower[:,:,1]
blue_channel = BerkeleyTower[:,:,2]

# use the following formula to obtain a gray scale image
gray_image = 0.2989 * red_channel + 0.5870 * green_channel + 0.1140 * blue_channel 

# display the result. 
imgplot = plt.imshow(gray_image, 'gray')

# following code is to save this converted image to local folder 
# It does not look like gray, but actually it is, 
# because it has only 2 dimensions.
img_path = r'C:\Users\ijimr\Zhen\ImageBank\GrayTower.jpg'
plt.imsave(img_path, gray_image)
```

```python
# How to generate an random image with float format? 
import numpy as np
import matplotlib.pyplot as plt

# create a random 2-D array with 100 by 100 size
noise  = np.random.random_sample((100,100))

# it is optional if you want to convert it into float32 format 
noise = noise.astype(np.float32)

# now, display the noise image
plt.imshow(noise, 'gray')

# Note that in order to avoid some potential syntax error, 
# we add letter r at the beginning of the quoted path
img_path = r'C:\Users\ijimr\Zhen\ImageBank\Nosie.jpg'

# Here is one method to save the image:
plt.imsave(img_path, noise)
```

```python
# How to generate an random image with uint8 format? 
import numpy as np
import matplotlib.pyplot as plt

# create a random 2-D array with 100 by 100 size
noise  = np.random.random_sample((100,100))* 255

# it is optional if you want to convert it into uint8 format 
noise = noise.astype(np.uint32)

# now, display the noise image
plt.imshow(noise, 'gray')

# Note that in order to avoid some potential syntax error, 
# we add letter r at the beginning of the quoted path
img_path = r'C:\Temp\Nosie.jpg'

# Here is one method to save the image:
plt.imsave(img_path, noise)
```

```python
# Histogram Equalization
import matplotlib.pyplot as plt
from skimage import exposure

# read the image from local folder useing plt.imread function 
Hawkes_Bay = plt.imread(r'C:\Users\ijimr\Zhen\ImageBank\Hawkes_Bay_NZ.jpg')

# use exposure.equalize_hist to improve the contrast of an impage 
Hawkes_Bay_eq = exposure.equalize_hist(Hawkes_Bay)

# display images in a row
plt.subplot(121), plt.imshow(Hawkes_Bay, 'gray')
plt.subplot(122), plt.imshow(Hawkes_Bay_eq, 'gray')
plt.show()
```

```python
# Practice for Image Thresholding
import cv2
import numpy as np
from matplotlib import pyplot as plt

tiger = plt.imread(r'C:\Users\ijimr\Zhen\ImageBank\tiger.jpg')

ret,thresh1 = cv2.threshold(tiger,127,255,cv2.THRESH_BINARY)


titles = ['Original Image','Binary Thresholding']
images = [tiger, thresh1]

for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])

plt.show()
```

```python
# Practice for Smoothing Images and Removing Noises
import cv2
import numpy as np
from matplotlib import pyplot as plt

polluted_circuit = plt.imread(r'C:\Users\ijimr\Zhen\ImageBank\circuitNoise.jpg')

clean_circuit = cv2.blur(polluted_circuit,(5,5))

plt.subplot(121),plt.imshow(polluted_circuit, 'gray'),plt.title('Original')
plt.subplot(122),plt.imshow(clean_circuit, 'gray'),plt.title('Noise_removal')
plt.show()
```
